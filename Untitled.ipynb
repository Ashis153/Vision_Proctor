```python
import json
import zipfile

# 1. Point to your zip file
zip_path = 'Online Proctoring System.coco.zip'

# 2. Open the zip "container"
with zipfile.ZipFile(zip_path, 'r') as z:
    # Let's list the files to make sure we find the right JSON path
    # Roboflow usually puts it in 'train/_annotations.coco.json'
    all_files = z.namelist()
    json_file_path = [f for f in all_files if f.endswith('.json')][0]
    
    print(f"Reading from: {json_file_path}")

    # 3. Open the JSON file *inside* the zip
    with z.open(json_file_path) as f:
        # We read the content and then decode it from bytes to utf-8 text
        data = json.loads(f.read().decode('utf-8'))

# Now your existing logic will work!
my_images_list = []
for img in data['images'][:5]:
    my_images_list.append({
        "id": img['id'],
        "file_name": img['file_name'],
        "height": img['height'],
        "width": img['width']
    })

print(json.dumps(my_images_list, indent=4))
```

```python
import json
import zipfile
import pandas as pd
import os

# --- CONFIGURATION ---
zip_path = 'Online Proctoring System.coco.zip'
extract_dir = 'proctor_dataset' # Folder where images will go
csv_output = 'proctor_labels.csv'

# 1. UNZIP EVERYTHING (Images + JSON)
print("Unzipping dataset... please wait.")
with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall(extract_dir)
print(f"Extraction complete! Files saved to: {extract_dir}")

# 2. LOCATE AND LOAD THE JSON
# Roboflow usually puts it in 'train/_annotations.coco.json'
json_path = os.path.join(extract_dir, 'train', '_annotations.coco.json')

with open(json_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 3. CREATE LOOKUP DICTIONARIES
image_map = {img['id']: img['file_name'] for img in data['images']}
category_map = {cat['id']: cat['name'] for cat in data['categories']}

# 4. PROCESS ANNOTATIONS INTO CSV FORMAT
rows = []
for ann in data['annotations']:
    # Get bbox: [x_min, y_min, width, height]
    bbox = ann['bbox']
    
    # FIX: Force everything to float then int to avoid 'str' + 'int' errors
    try:
        xmin = float(bbox[0])
        ymin = float(bbox[1])
        w    = float(bbox[2])
        h    = float(bbox[3])
        
        rows.append({
            'frame': image_map.get(ann['image_id'], "Unknown"),
            'xmin': int(xmin),
            'xmax': int(xmin + w), # Kaggle format needs max X
            'ymin': int(ymin),
            'ymax': int(ymin + h), # Kaggle format needs max Y
            'class_id': ann['category_id'],
            'label_name': category_map.get(ann['category_id'], "Unknown")
        })
    except (ValueError, TypeError):
        continue # Skip any broken data rows

# 5. SAVE AND PREVIEW
df = pd.DataFrame(rows)
df.to_csv(csv_output, index=False)

print("\n" + "="*30)
print(f"SUCCESS: {csv_output} created!")
print(f"Total Labels: {len(df)}")
print("="*30)
print(df.head())
```
